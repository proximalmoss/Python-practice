```python
import seaborn as sns
```

```python
import requests, certifi, io, pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
r = requests.get(url, verify=certifi.where())
r.raise_for_status()
df = pd.read_csv(io.StringIO(r.text))
```

```python
df.head()
```

Correlation with Heatmap

```python
df.corr(numeric_only=True)
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

Join Plot (Univariate Analysis)

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
#suggests that many people have tipped $2 and majority of the bill is btw $10-$20
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

Pair Plot (when more than two features)

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

Dist Plot

```python
sns.distplot(df['tip']) #gives in percentage
```

```python
sns.distplot(df['tip'],kde=False,bins=10) #if kde is false we get count
```

CATEGORICAL PLOTS

```python
#count plot
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
#bar plot
sns.barplot(y='total_bill',x='smoker',data=df)
```

```python
#box plot
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v') #giving percentile
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

```python
#Violin plot
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
