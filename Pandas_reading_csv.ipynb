```python
from io import StringIO, BytesIO
```

```python
data = ('col1,col2,col3\n'
        'x,y,1\n'
        'a,b,2\n'
        'c,d,3')
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data))
```

```python
#reading from specific columns
df=pd.read_csv(StringIO(data),usecols=lambda x:x.upper() in ['COL1','COL3'])
```

```python
#lambda not needed, just picking up column1 and column3
df
```

```python
df.to_csv('test.csv')
```

```python
#specifying columns datatype
data=('a,b,c,d\n'
      '1,2,3,4\n'
      '5,6,7,8\n'
      '9,10,11')
```

```python
print(data)
```

```python
df=pd.read_csv(StringIO(data),dtype=object)
df
```

```python
df['a']
```

```python
df['a'][1]
```

```python
df=pd.read_csv(StringIO(data),dtype={'b':int,'c':float,'a':'Int64'})
df
```

```python
#checking the datatypes
df.dtypes
```

```python
#making the first column as index
pd.read_csv(StringIO(data),index_col=0) #if we mention 1 then b and 2 then c will become index
```

```python
data=('a,b,c\n'
      '4,apple,bat,\n'
      '8,orange,cow')
```

```python
pd.read_csv(StringIO(data),index_col=False)
```

```python
#combining usecols and index_col
pd.read_csv(StringIO(data), usecols=['b','c'],index_col=False)
```

```python
#escape character
data= 'a,b\n"hello,\\"Bob\\",nice to see you",5'
pd.read_csv(StringIO(data),escapechar='\\')
```
